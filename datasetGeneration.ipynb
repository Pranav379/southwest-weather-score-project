```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import gdown
import io

# URL from Google Drive
url = 'https://drive.google.com/file/d/1TKoGKkqsHhJdJstk4fiQKO9Z2VY2GI0b/view?usp=sharing'

# Get the file ID from the URL
file_id = url.split('/')[-2]

# Use gdown to download and return the content
content = gdown.download(f'https://drive.google.com/uc?id={file_id}', quiet=False, fuzzy=True)

# Read the CSV
df = pd.read_csv(content)

print(df.shape)
print(df.head())
```
